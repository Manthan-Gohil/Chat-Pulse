```python
import pandas as pd
import re
```

```python
f = open('main.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[ap]m\s-\s'
# pattern = r'\d{1,2}\/\d{1,2}\/\d{2},\s\d{1,2}:\d{2}\s(?:AM|PM)'     
```

```python
messages = re.split(pattern, data)[1:]
len(messages)
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message':messages, 'message_date': dates})
# convert message_date type
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%Y, %I:%M %p - ')

df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
# separate users and messages
users = []
messages = []
for message in df['user_message']:
    entry = re.split(r'([\w\W]+?):\s', message)
    if entry[1:]: #user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['messsage'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
#making new column of year name from the date column
df['year'] = df['date'].dt.year
```

```python
f = open('stop_hinglish.txt','r')
stop_words = f.read()
```

```python
# remove stop words
words = []

for message in temp['messsage']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)
```

```python
from collections import Counter
y = pd.DataFrame(Counter(words).most_common(20))
```

```python
sns.barplot(x = y[0],y = y[1],data = y)
plt.xticks(rotation='vertical')
plt.show()
```

```python
period = []
for hour in df[['day_name','hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['period'] = period
```

```python
df.sample(5)
```

```python
plt.figure(figsize=(20,6))
sns.heatmap(df.pivot_table(index='day_name',columns='period',values='messsage',aggfunc='count').fillna(0))
plt.yticks(rotation='horizontal')
plt.show()
```
